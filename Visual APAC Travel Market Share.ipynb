```python
# packages to store and manipulate data
import numpy as np
import pandas as pd

# plotting tools
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Import Dataset
df = pd.read_csv('C:/Users/607791/Desktop/DS/Data Visualization/Week 6/APAC_Travel_Market3.csv')
df.head()
```

```python
tableau20 = [(31, 119, 180), (174, 199, 232), (255, 127, 14), (255, 187, 120),    
             (44, 160, 44), (152, 223, 138), (214, 39, 40), (255, 152, 150),    
             (148, 103, 189), (197, 176, 213), (140, 86, 75), (196, 156, 148),    
             (227, 119, 194), (247, 182, 210), (127, 127, 127), (199, 199, 199),    
             (188, 189, 34), (219, 219, 141), (23, 190, 207), (158, 218, 229)]    
    
for i in range(len(tableau20)):    
    r, g, b = tableau20[i]    
    tableau20[i] = (r / 255., g / 255., b / 255.)    
  
 
plt.figure(figsize=(14, 14/1.5))    
  
    
       
plt.ylim(0, 50)    
plt.xlim(1999, 2018)
   
plt.yticks(range(0, 50, 10), [str(x) + "%" for x in range(0, 50, 10)], fontsize=14)    
#plt.xticks(fontsize=14)    
     
for y in range(10, 50, 10):    
    plt.plot(range(2000, 2016), [y] * len(range(2000, 2016)), "--", lw=0.5, color="black", alpha=0.3)    
       
majors = ['Asia Pacific', 'Europe', 'North America', 'Other',    
          'Latin America', 'Middle East']    
  
for rank, column in enumerate(majors):       
    plt.plot(df.Year.values,    
            df[column.replace("\n", " ")].values,    
            lw=2.5, color=tableau20[rank])  
    
    y_pos = df[column.replace("\n", " ")].values[1]

    
   
    if column == "Asia Pacific":    
        y_pos += 0.5   
    elif column == "Europe":    
        y_pos += 1    
    elif column == "North America":    
        y_pos -= 1.25   
    elif column == "Other":    
        y_pos += 0.5    
    elif column == "Latin America":    
        y_pos += 1   
    elif column == "Middle East":    
        y_pos -= 1.5     
    plt.text(2015, y_pos, column, fontsize=14, color=tableau20[rank]) 
       
    
plt.text(2008, 52, "Percentage of APAC Travel Market Share"    
       " (2000-2016)", fontsize=22, ha="center")    
  

plt.text(2000, -6, "Data source: ...."    
       "\nAuthor: Parker Hall"    
       , fontsize=10)    
 
plt.savefig("percent-bachelors-degrees-women-usa.png") 
```
